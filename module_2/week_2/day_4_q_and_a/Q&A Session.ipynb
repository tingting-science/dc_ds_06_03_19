```python
def name_starts_with(drivers, substring):
    return_list = []
    for driver in drivers:
        if driver.name and driver.name.startswith(str(substring)):
            return_list.append(driver.name)
    if return_list:
        return return_list
    else:
        print(f"no names found starting with {substring}")
```

```python
def name_starts_with_generator(drivers, substring):
    for driver in drivers:
        if driver.name and driver.name.startswith(str(substring)):
            yield driver.name
#         print(f"no names found starting with {substring}")
```

```python
def find_driver_by_name(drivers, name):
    # write your code here
    for driver in drivers: 
        if driver.name == name: 
            return print(f"found {name}!")
    print("Sorry we couldn't find a driver with the name, {}! :(".format(name))
    return None
```

```python
#Misha's question
class Guitarist(object):
    """
    This is Guitarist's docstring
    """
    
    age = 50
    
    def __init__(self, name, role):
        """
        this is init
        """
        self.name = name
        self.role = role
        self.instrument_type = "Stringed Instrument"
    
    @classmethod
    def tune_instrument(cls):
        """
        more docstrings
        """
        print("Tune the strings!")
        
    def practice(self):
        print("Strumming the old 6 string!")

    @staticmethod
    def perform(new_name):
        print(f"Welcome to, New York, {new_name}!")
        
class Bass_Guitarist(Guitarist):
    
    def __init__(self):
        super().__init__(None, None)
        self.name = "Paul"
        self.role = "Guitarist"
        
    def practice(self):
        print("I play the Seinfeld Theme Song when I get bored")
        
    def perform(self):
        super().perform()
        print("Thanks for coming out!")
```
